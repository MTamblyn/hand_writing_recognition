```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
mu1 = 10
var1 = 1
sigma1 = np.sqrt(var1)
mu2 = 11
var2 = 1
sigma2 = np.sqrt(var2)
p = np.linspace(mu1 - 3*sigma1, mu1 + 3*sigma1, 100)
q = np.linspace(mu2 - 3*sigma2, mu2 + 3*sigma2, 100)
plt.plot(p, stats.norm.pdf(p, mu1, sigma1), label = r'$p(x)$')
plt.plot(q, stats.norm.pdf(q, mu2, sigma2), label = r'$q(x)$')
plt.legend()
plt.show()
```

```python
def kl_divergence(p, q):
    return [p[i] * np.log(p[i]/q[i]) for i in range(len(p))]
```

```python
plt.plot(kl_divergence(p, q))
plt.show()
```
