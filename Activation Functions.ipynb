```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#activation functions
def sigmoid(x, der = False):
    if der:
        return np.exp(-x)/(np.exp(-x)+1)**2
    else:
        return 1/(1+np.exp(-x))
    
def ReLu(x, der = False):
    if der:
        if x > 0:
            return 1
        elif x == 0:
            return np.nan
        else:
            return 0
    else:
        return np.maximum(0, x)
```

```python
x = np.linspace(-6, 6, 1000)
y = sigmoid(x)
```

```python
plt.plot(x,y)
plt.legend(["Sigmoid Function"])
plt.show()
```

```python
y_der = sigmoid(x, True)

plt.plot(x, y_der)
plt.legend(["Sigmoid Derivative"])
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y = ReLu(x)
```

```python
plt.plot(x,y)
plt.legend([" ReLu Function"])
plt.show()
```

```python
y = [ReLu(i, True) for i in x] 
plt.plot(x, y)
plt.legend(["ReLu Derivative"])
plt.show()
```
