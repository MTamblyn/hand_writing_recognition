```python
import numpy as np
from scipy.stats import t, norm
import matplotlib.pyplot as plt
%matplotlib inline

rv = t(df=20, loc=0, scale=1)
x = np.linspace(rv.ppf(0.0001), rv.ppf(0.9999), 100)
y = rv.pdf(x) 

plt.xlim(-5,5)
plt.plot(x,y)
plt.show()
```

```python


x1 = np.linspace(-5, 5, 100)
pdf1 = norm.pdf(x1)
x2 = np.linspace(-5, 5, 100)
pdf2 = norm.pdf(x2)

from matplotlib import pyplot as plt
plt.plot(x1, pdf1, label="p(x)")
plt.plot(x2, pdf2, label='q(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0,np.pi*2,np.pi*2/100)
Y = np.sin(X)

slope_Y = np.diff(Y)/np.diff(X)

plt.plot(X,Y)
plt.plot(X[:-1],slope_Y)
plt.show()
```
